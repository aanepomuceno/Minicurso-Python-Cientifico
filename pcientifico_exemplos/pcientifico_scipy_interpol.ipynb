```python
import numpy as np
from scipy. interpolate import interp1d
from scipy. interpolate import interp2d
import matplotlib . pyplot as plt
%matplotlib inline
```

**1. Interpolação em 1D**

```python
#vamos tomar como exemplo A*exp(-kx)*cos(2*pi*nu*x)
A, nu, k = 10,4,2
def f(x, A, nu , k):
    return A * np.exp(-k*x) * np.cos (2* np.pi * nu * x)
```

```python
xmax  = 0.5 
x = np. linspace (0, xmax , 8)
y = f(x, A, nu , k)
```

```python
#gerar as interpolacoes
f_linear  = interp1d(x, y)
f_nearest = interp1d(x, y, kind='nearest')
f_cubic   = interp1d(x, y, kind='cubic')
```

```python
x2 = np. linspace (0, xmax , 100)
plt.plot(x, y, 'ko', label ='Dados')
plt.plot(x2, f(x2, A, nu, k), 'k',label ='Exato')
plt.plot(x2, f_nearest(x2), 'b--',label ='Nearest')
plt.plot(x2, f_linear(x2), 'y',label ='Linear')
plt.plot(x2, f_cubic (x2), 'r:',label ='Cubic')
plt.legend(frameon=False,loc='upper right')
#plt.savefig('../pcientifico_imagens/plot_29.png',dpi=300)
```

**2. Interpolação em 2D**

```python
x = np. linspace (0, 4, 13)
y = np.array ([0, 2, 3, 3.5, 3.75, 3.875, 3.9375, 4])
X, Y = np.meshgrid (x, y)
Z = np.sin(np.pi*X/2) * np.exp(Y/2)
```

```python
x2 = np.linspace (0, 4, 65)
y2 = np.linspace (0, 4, 65)
f = interp2d(x, y, Z, kind='cubic')
Z2 = f(x2, y2)
```

```python
fig, ax = plt.subplots(nrows =1, ncols=2,figsize=(6,5))
ax[0].pcolormesh(X, Y, Z)
X2, Y2 = np.meshgrid(x2, y2)
ax[1].pcolormesh(X2, Y2, Z2)
```
